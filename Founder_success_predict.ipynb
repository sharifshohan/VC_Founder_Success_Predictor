```python
def remove_duplicate_founders(dataframe):
    """
    Remove duplicate rows from a DataFrame based on the 'founder_linkedin_url' field.

    Args:
        dataframe (pd.DataFrame): The input DataFrame containing a 'founder_linkedin_url' column.

    Returns:
        pd.DataFrame: A new DataFrame with duplicate rows removed based on 'founder_linkedin_url'.
    """
    # Drop duplicate rows based on 'founder_linkedin_url'
    unique_df = dataframe.drop_duplicates(subset='linkedin_url', keep='first')

    # Reset the index for the unique DataFrame
    unique_df.reset_index(drop=True, inplace=True)

    return unique_df
```

```python
import json

def get_json_string_key_entries(dataframe):
    key_entries = list()
    for json_string in dataframe['linkedin_json'].tolist():
        try:
            key_entries += list(json.loads(json_string).keys())
        except json.JSONDecodeError:
            # Handle the error, e.g., log the error or skip the invalid JSON string
            print(f"Invalid JSON string: {json_string}")
    return set(key_entries)

# Example usage:
x = get_json_string_key_entries(filtered_merged_successfull_founder_company_df)
x
```

```python
from scholarly import scholarly
import time

# Function to get H-index from Google Scholar
def get_h_index(name):
    try:
        search_query = scholarly.search_author(name)
        author = next(search_query, None)
        if author:
            scholarly.fill(author, sections=['indices'])
            return author.get('hindex', None)
        else:
            return None
    except Exception as e:
        print(f"Error fetching H-index for {name}: {e}")
        return None
```

```python
import json
import pandas as pd

# Function to extract specific fields from JSON string
def extract_fields(json_str):
    try:
        data = json.loads(json_str)
        lead = data.get("lead", {})
        extracted = {
            "full_name": lead.get("full_name"),
            "industry": lead.get("industry"),
            "location": lead.get("location"),
            "location_state": lead.get("location_state"),
            "location_country": lead.get("location_country"),
            "job_title": lead.get("job_title"),
        }
        
        educations = lead.get("educations", [])
        if educations:
            extracted["fields_of_study"] = educations[0].get("fields_of_study")
            extracted["degree"] = educations[0].get("degree")
            extracted["university_name"] = educations[0].get("university_name")
        else:
            extracted["fields_of_study"] = None
            extracted["degree"] = None
            extracted["university_name"] = None
        
        return extracted
    except json.JSONDecodeError:
        return None

```

```python
# Extract fields and create a new DataFrame
extracted_data = [extract_fields(row) for row in filtered_merged_successfull_founder_company_df['linkedin_json']]
extracted_data = [row for row in extracted_data if row is not None]
result_df = pd.DataFrame(extracted_data)
#result_df['h_index'] = result_df['full_name'].apply(get_h_index)
# Print the result DataFrame
print(result_df)

# Concatenate the original df and result_df side by side
combined_filtered_merged_successfull_founder_company_df = pd.concat([filtered_merged_successfull_founder_company_df, result_df], axis=1)
```

```python
#Ranking the university and putting it in the feature 

# Load your dataframe (assuming it is named df)
df = combined_filtered_merged_successfull_founder_company_df

# Load the THE rankings
the_rankings = pd.read_csv('Data/TIMES_WorldUniversityRankings_2024.csv')

# Ensure the 'rank' column is numeric
the_rankings['rank'] = pd.to_numeric(the_rankings['rank'], errors='coerce')

# Drop rows with NaN ranks (optional, depending on how you want to handle them)
the_rankings.dropna(subset=['rank'], inplace=True)

# Convert the 'rank' column to integers
the_rankings['rank'] = the_rankings['rank'].astype(int)

# Add a column for the category
def categorize(rank):
    if rank <= 100:
        return 'Top'
    elif rank <= 500:
        return 'Medium'
    else:
        return 'Bottom'

the_rankings['Category'] = the_rankings['rank'].apply(categorize)

# Display the first few rows to check
print(the_rankings.head())

```
